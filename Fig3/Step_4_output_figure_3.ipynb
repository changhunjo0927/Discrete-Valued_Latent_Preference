```python
import scipy.io as io

loaded_1 = io.loadmat('data_1') # theta = 0 case
loaded_2 = io.loadmat('data_2') # theta = 0.15 case
loaded_3 = io.loadmat('data_3') # theta = 0.3 case

#print(loaded['obs'])
#print(loaded['max'])
#print(loaded['L1'])
#print(loaded['obs_mean'])
#print(loaded['max_mean'])
#print(loaded['L1_mean'])

import matplotlib.pylab as plt
import matplotlib as mpl
import pandas as pd

plt.rc('font', family='serif', serif='Times')
plt.rc('text', usetex=True)
plt.rc('xtick', labelsize=8)
plt.rc('ytick', labelsize=8)
plt.rc('axes', labelsize=8)
plt.rc('axes', linewidth=0.5)

mpl.rcParams['patch.linewidth']=0.7

width = 3.27
height = width / 1.618


fig, ax = plt.subplots()
fig.subplots_adjust(left=.16, bottom=.18, right=.99, top=.97)
fig.set_size_inches(width, height)

p_opt = 0.32747974577838707

x_1 = loaded_1['obs_mean'][0]
y_1 = loaded_1['max_mean'][0]
z_1 = loaded_1['L1_mean'][0]

x_2 = loaded_2['obs_mean'][0]
y_2 = loaded_2['max_mean'][0]
z_2 = loaded_2['L1_mean'][0]

x_3 = loaded_3['obs_mean'][0]
y_3 = loaded_3['max_mean'][0]
z_3 = loaded_3['L1_mean'][0]


max_x_1, max_y_1 = zip(*sorted(zip(x_1,y_1)))
L1_x_1, L1_z_1 = zip(*sorted(zip(x_1,z_1)))

max_x_2, max_y_2 = zip(*sorted(zip(x_2,y_2)))
L1_x_2, L1_z_2 = zip(*sorted(zip(x_2,z_2)))

max_x_3, max_y_3 = zip(*sorted(zip(x_3,y_3)))
L1_x_3, L1_z_3 = zip(*sorted(zip(x_3,z_3)))

plt.xlim(0.07,0.49)

plt.plot(max_x_1, max_y_1, color='orange', linewidth=0.6, marker = 'o', markersize = 1.2)
plt.plot(max_x_2, max_y_2, color='darkgray', linewidth=0.6, marker = 's', markersize = 1.2)
plt.plot(max_x_3, max_y_3, color='black', linewidth=0.6, marker = '^', markersize = 1.2)
plt.axvline(x=p_opt, color='r', linestyle=':', linewidth=0.8)
plt.legend([r'$\theta=0$',r'$\theta=0.15$',r'$\theta=0.3$',r'$p^*_{\gamma}$'], fontsize=8, loc='upper right')
plt.xlabel(r'$p$')
plt.ylabel(r'$||\hat{R}-R||_{\max}$')
plt.savefig('Figure_3.eps', format='eps', dpi=1000)
plt.show()

```
